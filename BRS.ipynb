# POPULARITY BASED RECOMMENDER SYSTEM

```python
ratings.merge(books, on='ISBN').shape
```

```python
ratings_with_name = ratings.merge(books, on='ISBN')
```

```python
rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].count().reset_index()
```

```python
rating_df.rename(columns={'Book-Rating':'num_rating'},inplace=True)
rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
popular_df = rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df=popular_df[popular_df['num_rating']>=250].sort_values('avg_rating',ascending=False).head(50)
```

```python
top_50_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Publisher','Image-URL-M','num_rating','avg_rating']]
```

# Collaborative Filtering Based Recommender System

```python
x = ratings_with_name.groupby('User-ID')['Book-Rating'].count() > 200
actual_raters = x[x].index
```

```python
actual_raters
```

```python
filtered_raters = ratings_with_name[ratings_with_name['User-ID'].isin(actual_raters)]
```

```python
y=filtered_raters.groupby('Book-Title')['Book-Rating'].count()>= 50
famous_book = y[y].index
```

```python
famous_book
```

```python
final_ratings = filtered_raters[filtered_raters['Book-Title'].isin(famous_book)]
final_ratings
```

```python
pt = final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
pt
```

```python
pt.fillna(0, inplace=True)
pt
```

```python
# !pip install scikit-learn
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(pt)
```

```python
similarity
```

```python
def recommend(book):
    index = np.where(pt.index == book)[0][0]
    similar_item = sorted(list(enumerate(similarity[index])),key = lambda x:x[1],reverse=True)[1:6]

    for i in similar_item:
        print(pt.index[i[0]])
```

```python
recommend('The Notebook')
```
